```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#load the data
data = pd.read_csv(r'C:\Users\Dell\Downloads\HR Data.csv')
data.head()
```

```python
# dropping passed columns
data.drop(["EmployeeCount"], axis = 1, inplace = True)
```

```python
#display
data.head()
```

```python
#giving the columns new names
data.rename(columns = {'YearsWithCurrManager': 'YearsWithCurrentManager','TotalWorkingYears': 'YearsOfExperience','EmployeeNumber': 'EmployeeID'}, inplace = True)
```

```python
#display
data.head()
```

```python
#eliminating redundant values
data = data.drop_duplicates()
```

```python
#display
data.head()
```

```python
#checking if dataset has any NaN values
has_nan = data.isnull().values.any()

if has_nan:
    print("There are NaN values in the Dataset.")
else:
    print("There are no NaN values in the Dataset.")
```

```python
#eliminating rows with nan values
data = data.dropna() 
```
